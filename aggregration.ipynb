```python
import pandas as pd
import numpy as np
```

```python
states = ["Alabama",
"Alaska",
"Arizona",
"Arkansas",
"California",
"Colorado",
"Connecticut",
"Delaware",
"Florida",
"Georgia",
"Hawaii",
"Idaho",
"Illinois",
"Indiana",
"Iowa",
"Kansas",
"Kentucky",
"Louisiana",
"Maine",
"Maryland",
"Massachusetts",
"Michigan",
"Minnesota",
"Mississippi",
"Missouri",
"Montana",
"Nebraska",
"Nevada",
"New Hampshire",
"New Jersey",
"New Mexico",
"New York",
"North Carolina",
"North Dakota",
"Ohio",
"Oklahoma",
"Oregon",
"Pennsylvania",
"Rhode Island",
"South Carolina",
"South Dakota",
"Tennessee",
"Texas",
"Utah",
"Vermont",
"Virginia",
"Washington",
"West Virginia",
"Wisconsin",
"Wyoming"]
```

```python
df_list = []

for state in states:
    df = pd.read_csv(f'tbcov_reduce\\us_states\\us_{state}.csv')
    
    # convert str type to datetime
    df['date_time'] = pd.to_datetime(df['date_time'])
    
    df_pivot = df.pivot_table(index=df['date_time'].dt.date, columns='sentiment_label', aggfunc='size', fill_value=0)
    
    # Cleaning
    df_pivot = df_pivot.reset_index().rename_axis(columns=None)
    df_pivot = df_pivot.rename(
        columns={
            -1: 'neg',
            1: 'pos',
            0: 'neutral'
        }
    )

    # Create columns
    df_pivot = df_pivot.assign(sum=df_pivot.iloc[:, 1:].sum(axis=1))
    df_pivot['prop_neg'] = df_pivot['neg']/df_pivot['sum']
    df_pivot['prop_pos'] = df_pivot['pos']/df_pivot['sum']
    df_pivot['prop_neutral'] = df_pivot['neutral']/df_pivot['sum']
    df_pivot['state'] = state

    df_list.append(df_pivot)

df = pd.concat(df_list)

df.to_csv('us_states_tweet_count.csv', index=False)
```
